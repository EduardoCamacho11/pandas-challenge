# PyCity Schools Analysis

```python
import pandas as pd
from pathlib import Path
```

```python
school_csv_path = Path("../Resources/schools_complete.csv")
student_csv_path = Path("../Resources/students_complete.csv")
```

```python
school_data = pd.read_csv(school_csv_path)
student_data = pd.read_csv(student_csv_path)
```

```python
school_data_complete = pd.merge(student_data, school_data, how= "left", on="school_name")
school_data_complete.head()
```

## District Summary

```python
school_count = school_data_complete['school_name'].nunique()
print(school_count)
```

```python
student_count = school_data_complete['Student ID'].nunique()
student_count
```

```python
total_budget = school_data_complete['budget'].unique().sum()
total_budget
```

```python
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
passing_math_count = school_data_complete[(school_data_complete['math_score'] >=70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count)*100
passing_math_percentage
```

```python
passing_reading_count = school_data_complete[(school_data_complete['reading_score'] >=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count)*100
passing_reading_percentage
```

```python
passing_math_reading_count = school_data_complete[
    (school_data_complete['math_score'] >=70) & (school_data_complete['reading_score'] >=70)].count() ['student_name']
overall_passing_rate = passing_math_reading_count / float(student_count) * 100
overall_passing_rate
```

```python
district_summary = pd.DataFrame(
    {'Total Unique Schools': [school_count], 
     'Total Students': [student_count], 
    'Total Budget': [total_budget], 
     'Average Math Score': [average_math_score],
     'Average Reading Score': [average_math_score], 
     'Passing Math Percetange': [passing_math_percentage],
     'Passing Reading Percentage': [passing_reading_percentage],
     'Overall Passing Percentage': [overall_passing_rate]})

district_summary

```

## School Summary

```python
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
per_school_count = school_data.set_index(["school_name"])["size"]
per_school_count
```

```python
per_school_budget = school_data.set_index(["school_name"])["budget"]
per_school_budget
```

```python
per_school_capita = per_school_budget / per_school_count
per_school_capita
```

```python
per_school_math = school_data_complete.groupby(["school_name"])["math_score"].mean()
per_school_math
```

```python
per_school_reading = school_data_complete.groupby(["school_name"])["reading_score"].mean()
per_school_reading
```

```python
students_passing_math = school_data_complete[(school_data_complete['math_score'] >=70)]
school_students_passing_math = students_passing_math.groupby("school_name")["student_name"].count()
school_students_passing_math
```

```python
students_passing_reading = school_data_complete[(school_data_complete['reading_score'] >=70)]
school_students_passing_reading = students_passing_reading.groupby("school_name")["student_name"].count()
school_students_passing_reading
```

```python
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
per_school_passing_math = school_students_passing_math / per_school_count * 100
per_school_passing_math
```

```python
per_school_passing_reading = school_students_passing_reading / per_school_count * 100
per_school_passing_reading
```

```python
overall_passing_rate = school_students_passing_math_and_reading / per_school_count * 100
overall_passing_rate
```

```python
per_school_math
```

```python
per_school_summary = pd.DataFrame(
    {'School Name': school_types, 
     'Total Students': per_school_count, 
     'Total School Budget': per_school_budget, 
     'Per Student Budget': per_school_capita, 
     'Average Math Score': per_school_math,
     'Average Reading Score': per_school_reading, 
     'Percentage Passing Math': per_school_passing_math,
     'Percentage Passing Reading': per_school_passing_reading, 
     'Percentage Overall Passing': overall_passing_rate})


per_school_summary
```

## Scores by School Spending

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_df = per_school_summary.copy()
```

```python
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budge"],
                                                             spending_bins, labels, include_lowest=True)
school_spending_df["Spending Ranges (Per Student)"] = school_spending_df["Spending Ranges (Per Student)"].astype(str)
school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Percentage Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Percentage Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Percentage Overall Passing"].mean()

spending_summary = pd.DataFrame(
    {'Spending Average Math Score': spending_math_scores,
     'Spending Average Reading Score': spending_reading_scores,
     'Spending Passing Math Percentage': spending_passing_math,
     'Spending Passing Reading Percentage': spending_passing_reading,
     'Overall Spending Passing Percentage': overall_passing_spending})
spending_summary 
```

## Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_size_df = per_school_summary.copy()
```

```python
school_size_df["School Size"] = pd.cut(school_size_df["Total Students"],
                                                      size_bins, labels,
                                                      include_lowest=True)
school_size_df["School Size"] = school_size_df["School Size"].astype(str)
school_size_df
```

```python
size_math_scores = school_size_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_size_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_size_df.groupby(["School Size"])["Percentage Passing Math"].mean()
size_passing_reading = school_size_df.groupby(["School Size"])["Percentage Passing Reading"].mean()
size_overall_passing = school_size_df.groupby(["School Size"])["Percentage Overall Passing"].mean()

size_summary = pd.DataFrame(
    {'Average Math Scores': size_math_scores,
     'Average Reading Scores': size_reading_scores,
    '% Passing Math': size_passing_math,
    '% Passing Reading':size_passing_reading,
    '% Overall Passing': size_overall_passing})

size_summary
```
